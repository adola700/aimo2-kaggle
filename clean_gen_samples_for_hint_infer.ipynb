```python
def extract_boxed_texts(text):
    text = text.strip("\n")
    pattern = r'\\boxed{((?:[^{}]|{[^{}]*})+)}'
    matches = re.findall(pattern, text)
    if not matches:
        return -1000
    # ans = []
    # for content in matches:
    #     if content.isdigit():
    #         num = int(content)
    #         ans.append(num % 1000)

    return matches[-1]
```
