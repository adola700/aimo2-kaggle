```python
def extract_boxed_texts(text):
    text = text.strip("\n")
    pattern = r'\\boxed{((?:[^{}]|{[^{}]*})+)}'
    matches = re.findall(pattern, text)
    if not matches:
        return -1000
    # ans = []
    # for content in matches:
    #     if content.isdigit():
    #         num = int(content)
    #         ans.append(num % 1000)

    return matches[-1]
```

```python
ga = list(check_df["gen_answers"]) 
a = list(check_df["answer"])
```

```python
from tqdm import tqdm
import math

corrects = []
for i in tqdm(range(math.ceil(len(a)/500))):
    start = 500*(i)
    end = 500*(i+1)
    correct_batch = parallelize(verify_correctness, ga[start:end], a[start:end])
    corrects.extend(correct_batch)
```

```python
corrects = [correct[0] for correct in corrects]
```

```python
pd.Series(corrects).value_counts()
```

```python
check_df["corrects"] = corrects
```

```python
check_df["corrects"] = check_df["corrects"].apply(lambda x: int(x) if (x == '0' or x == '1') else -1)
```
